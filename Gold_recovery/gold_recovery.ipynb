### Вычисление эффективности обогащения для чернового концентрата

Вычислим эффективность обогащения (Recovery) для этапа флотации (т.е. для признака ```rougher.output.recovery```)

````python
# Вычилсление эффективности обогащения для этапа
# c - доля золота в концентрате после флотации/очистки
# f - доля золота в сырье/концентрате до флотации/очистки
# t - доля золота в отвальных хвостах после флотации/очистки

def recovery(c,f,t):
    return c*(f-t)/(f*(c-t))*100
````

````python
recovery_rougher = recovery(data_train['rougher.output.concentrate_au'], data_train['rougher.input.feed_au'], 
                            data_train['rougher.output.tail_au'])

print('MAE между расчетом и значением из таблицы: ', mean_absolute_error(data_train['rougher.output.recovery'],recovery_rougher))
````

Средняя абсолютная ошибка пренебрежимо мала, следовательно, формула для вычислений верна.

## Модель

### Функция для вычисления sMAPE

````python
def wrong_way_smape(model, features, target):
    predicted = model.predict(features)
    predicted = pd.Series(predicted)
    target = pd.Series(target)
    target.index = range(0, len(target))
    values = 0
    for i in range(0, len(target)):
        values = values + 2 * (abs(target[i] - predicted[i]))/(abs(target[i]) + abs(predicted[i]))
    score = values*100/target.count()
    return score
````

````python
def smape(model, features, target):
    predicted = model.predict(features)
    predicted = pd.Series(predicted)
    target = pd.Series(target)
    target.index = range(0, len(target))
    values = 2 * (abs(target - predicted))/(abs(target) + abs(predicted))
    score = values.sum()*100/values.count()
    return score
````

#### Нахождение rougher.output.recovery

Разобьем тренировочную выборку на цели и признаки для нахождения rougher.output.recovery.  
Для прогнозирования этого параметра нам понадобятся только признаки, относящиеся к этапам до первичной очистки (rougher.xxx.yyy)

````python
target_train = data_train['rougher.output.recovery']
features_train = data_train.drop(['date', 'rougher.output.recovery'], axis=1)

# Отберем признаки, относящиеся к этапу Rougher
features_train = features_train.filter(like='rougher.', axis=1)

# Отберем признаки, присутствующие только в тестовом датасете
cols_to_add = data_test.drop([ 'rougher.output.recovery'], axis=1).filter(like='rougher.', axis=1).columns
features_train = features_train[cols_to_add]

# Запишем признаки и таргеты для DummyRegressor
target_dummy_rougher = target_train
features_dummy_rougher = features_train
````

Рассмотрим несколько моделей и оценим их качество.

````python
q = 5
model = LinearRegression()
scores = cross_val_score(model, features_train, target_train, cv=q, scoring=smape)
print(f'Линейная регрессия, sMAPE для чернового концентрата {sum(scores)/q:.2f}%')
````

````python
best_result = 100
best_depth = 0
for depth in range(1, 10):
    model = DecisionTreeRegressor(random_state=12345, max_depth=depth)
    scores = cross_val_score(model, features_train, target_train, cv=5, scoring=smape)
    result = sum(scores)/5
    if result < best_result:
        best_result = result
        best_depth = depth

#print("sMAPE наилучшей модели:", best_result, "Глубина дерева:", best_depth)

print(f'Дерево решений, sMAPE для чернового концентрата {best_result:.2f}%, глубина дерева {best_depth}')
````

````python
best_result = 100
best_est = 0
for est in range(10, 51, 10):
    model = RandomForestRegressor(random_state=12345, n_estimators=est, max_depth=best_depth)
    scores = cross_val_score(model, features_train, target_train, cv=5, scoring=smape)
    result = sum(scores)/5
    if result < best_result:
        best_result = result
        best_est = est

#print("RMSE наилучшей модели на валидационной выборке:", best_result, "Количество деревьев:", best_est, "Максимальная глубина:", depth)

print(f'Случайный лес, sMAPE для чернового концентрата {best_result:.2f}%, количество деревьев {best_est}')
````
